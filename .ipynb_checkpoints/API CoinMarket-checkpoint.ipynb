```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'50', ## initial was 5000
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': '93f89d66-2f23-4f0e-89aa-64234c8e410e',
}

session = Session()
session.headers.update(headers)

try:
    response = session.get(url, params=parameters)
    data = json.loads(response.text)
    print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
    print(e)
```

### DATA CLEANING

```python
pd.set_option("display.float_format", lambda x: "%.5f" % x)
```

```python
## from the data we pulled we had many duplicates like Bitcon 
df3=df2.groupby("name", sort=False)[["quote.USD.percent_change_1h","quote.USD.percent_change_24h","quote.USD.percent_change_7d","quote.USD.percent_change_30d","quote.USD.percent_change_90d"]].mean()
df3
```

```python
df4=df3.stack()
df4
```

```python
type(df4)
## it changes when you stack it from Dataframe to Series 
```

```python
## we need to convert it to a dataframe
df5= df4.to_frame(name="percent_change")
df5
```

```python
df5.count()
```

```python
# we are setting it to rows an assigning an index for visualization 

index = pd.Index(range(250))

df6 =df5.reset_index()
df6
```

```python
## we are renaming the column levels
df7= df6.rename(columns={"level_1":"Time_change"})
df7
```

```python
## we need to shorten some rows 
df7["Time_change"] = df7["Time_change"].replace(["quote.USD.percent_change_1h","quote.USD.percent_change_24h","quote.USD.percent_change_7d","quote.USD.percent_change_30d","70d"],["1h","24h","7d","30d","90d"])
df7
```

### Visualize Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.catplot(x="Time_change", y="percent_change", hue="name", data=df7, kind="point")
```
